Steady State Concentration

$$ C_s = \frac{D}{\left(\frac{\ln(2)}{t_{1/2}}\right) \times T_I} $$

Where:
- $C_s$ is the steady-state concentration result
- $D$ is the dose of the medication
- $\ln(2)$ is the natural logarithm of 2
- $t_{1/2}$ is the elimination half-life of the medication
- $T_I$ is the dosing interval

$$
C_s = \frac{D}{\left(\frac{\ln(2)}{t_{1/2}}\right) \times T_I}
$$

Given:
$$
D = 0.5 \, \text{mg} \quad (\text{dose})
$$
$$
T_I = 3 \, \text{hours} \quad (\text{dosing interval})
$$
$$
t_{1/2} = 3 \, \text{hours} \quad (\text{half-life})
$$

$$
C_s = 0.5 \times \left(\frac{\ln(2)}{3}\right) \times 3
$$

$$
C_s \approx \frac{0.5}{\frac{0.693}{3}} \approx \frac{0.5}{0.231} \approx 2.16 \, \text{mg}
$$


```python
# Imports
# Third-party
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Concentration level
def calculate_concentration(dose, half_life, dosing_interval, time):
    k = np.log(2) / half_life
    steady_state_concentration = dose / (k * dosing_interval)
    concentration = steady_state_concentration * (1 - np.exp(-k * time))
    return concentration

def calculate_concentration_with_timing(dose, half_life, dosing_interval, time, timing_offset):
    adjusted_time = time - timing_offset
    k = np.log(2) / half_life
    steady_state_concentration = dose / (k * dosing_interval)
    concentration = steady_state_concentration * (1 - np.exp(-k * adjusted_time))
    return concentration
```

```python
# Parameters for Roxanol (morphine)
roxanol_dose = 3  # mg
roxanol_half_life = 3  # hours
dosing_interval = 3 # hours

# Parameters for Hydromorphone
hydromorphone_dose = 3  # mg
hydromorphone_half_life = 2  # hours
hydromorphone_dosing_interval = 3  # hours (adjust as needed)

# Time points for the chart (in hours)
time_points = np.arange(0, 24, 0.1)

# Calculate concentrations
roxanol_concentration = [calculate_concentration(roxanol_dose, roxanol_half_life, dosing_interval, t) for t in time_points]
hydromorphone_concentration = [calculate_concentration(hydromorphone_dose, hydromorphone_half_life, hydromorphone_dosing_interval, t) for t in time_points]

# Plotting
plt.plot(time_points, roxanol_concentration, label=f'Roxanol ({dosing_interval}hr interval)')
plt.plot(time_points, hydromorphone_concentration, label=f'Hydromorphone ({hydromorphone_dosing_interval}hr interval)')

# Add labels and title
plt.xlabel('Time (hours)')
plt.ylabel('Concentration (mg)')
plt.title('Comparison of Roxanol and Hydromorphone Concentrations')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Parameters for Roxanol (morphine)
roxanol_dose = 3  # mg
roxanol_half_life = 3  # hours
dosing_interval = 3 # hours

# Parameters for Hydromorphone
hydromorphone_dose = 3  # mg
hydromorphone_half_life = 2  # hours
hydromorphone_dosing_interval = 3  # hours (adjust as needed)

# Parameters for Lorazepam (Intensol)
lorazepam_dose = 1  # mg (example dose)
lorazepam_half_life = 12  # hours
lorazepam_dosing_interval = 4  # hours (example dosing interval)

# Time points for the chart (in hours)
time_points = np.arange(0, 24, 0.1)

# Calculate concentrations
roxanol_concentration = [calculate_concentration(roxanol_dose, roxanol_half_life, dosing_interval, t) for t in time_points]
hydromorphone_concentration = [calculate_concentration(hydromorphone_dose, hydromorphone_half_life, hydromorphone_dosing_interval, t) for t in time_points]
lorazepam_concentration = [calculate_concentration(lorazepam_dose, lorazepam_half_life, lorazepam_dosing_interval, t) for t in time_points]

# Plotting
plt.plot(time_points, roxanol_concentration, label=f'Roxanol ({dosing_interval}hr interval)')
plt.plot(time_points, hydromorphone_concentration, label=f'Hydromorphone ({hydromorphone_dosing_interval}hr interval)')
plt.plot(time_points, lorazepam_concentration, label=f'Lorazepam ({lorazepam_dosing_interval}hr interval)')

# Add labels and title
plt.xlabel('Time (hours)')
plt.ylabel('Concentration (mg)')
plt.title('Comparison of Roxanol, Hydromorphone, and Lorazepam Concentrations')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Assuming a dose of 1 mg, half-life of 12 hours, and dosing interval of 6 hours for Lorazepam
lorazepam_dose = 2
lorazepam_half_life = 12
lorazepam_dosing_interval = 3

# Time points for the chart (in hours)
time_points = np.arange(0, 72, 0.1)

# Calculate concentrations with different timing offsets
lorazepam_concentration_on_time = [calculate_concentration_with_timing(lorazepam_dose, lorazepam_half_life, lorazepam_dosing_interval, t, 0) for t in time_points]
lorazepam_concentration_early = [calculate_concentration_with_timing(lorazepam_dose, lorazepam_half_life, lorazepam_dosing_interval, t, -1) for t in time_points]
lorazepam_concentration_late = [calculate_concentration_with_timing(lorazepam_dose, lorazepam_half_life, lorazepam_dosing_interval, t, 1) for t in time_points]

# Plotting
plt.plot(time_points, lorazepam_concentration_on_time, label='On time')
plt.plot(time_points, lorazepam_concentration_early, label='One hour early')
plt.plot(time_points, lorazepam_concentration_late, label='One hour late')

# Add labels and title
plt.xlabel('Time (hours)')
plt.ylabel('Concentration (mg)')
plt.title('Effect of Timing on Lorazepam Concentration')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
