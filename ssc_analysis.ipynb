```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from src.calculator import MedicationProfile, ConcentrationCalculator
from src.visualization import ConcentrationVisualizer
```

```python
# Initialize our calculator and visualizer
calculator = ConcentrationCalculator()
visualizer = ConcentrationVisualizer(calculator)
```

```python
# Create medication profiles
# These parameters can be easily modified for different scenarios
roxanol = MedicationProfile(
    name='Roxanol',
    dose=3,  # mg
    half_life=3,  # hours
    dosing_interval=3  # hours
)

hydromorphone = MedicationProfile(
    name='Hydromorphone',
    dose=3,  # mg
    half_life=2,  # hours
    dosing_interval=3  # hours
)

lorazepam = MedicationProfile(
    name='Lorazepam',
    dose=1,  # mg
    half_life=12,  # hours
    dosing_interval=3  # hours
)

# Add medications to calculator for comparison
calculator.add_medication(roxanol)
calculator.add_medication(hydromorphone)
calculator.add_medication(lorazepam)
```

```python
# Plot comparison of all medications
visualizer.plot_concentrations(
    time_range=(0, 24),
    title='Comparison of Roxanol, Hydromorphone, and Lorazepam Concentrations'
)
```

```python
# Create a new calculator/visualizer for timing analysis
timing_calculator = ConcentrationCalculator()
timing_visualizer = ConcentrationVisualizer(timing_calculator)

# Add just Lorazepam for timing analysis
timing_calculator.add_medication(lorazepam)

# Plot timing variations for Lorazepam
timing_visualizer.plot_timing_variations(
    medication=lorazepam,
    time_range=(0, 72)
)
```
